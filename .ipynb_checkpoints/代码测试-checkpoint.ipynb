# 利用广度优先法找出相近词

```python
from collections import defaultdict
```

```python
# 判断一个词是否是动词，是的话返回True
def is_verb(word):
    for word, flat in pseg.cut(word):
        if flat != 'v': return False
        
    return True
```

```python
# 原方法
def get_related_words(initial_words, model, topn):
    start = time.time()
    unseen = initial_words.split()
    
    seen = defaultdict(int)
    
    max_size = 400
    
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
                   
        new_expanding = [w for w, s in model.most_similar(node, topn=topn) if is_verb(w)]
        
        
        unseen += new_expanding
        
        seen[node] += 1
    
    end = time.time()
    print(end - start)
    return seen
```

```python
# 修改后
def get_related_words(initial_words, model, topn):
    start = time.time()
    unseen = initial_words.split()
    
    seen = defaultdict(int)
    
    max_size = 400
    
    # 建邻表
    adj = defaultdict(list)
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
        
#         print(node, seen)
        
        if node in seen:
            new_expanding = adj[node]
        else:            
            new_expanding = [w for w, s in model.most_similar(node, topn=topn) if is_verb(w)]
        
        adj[node] = new_expanding
        
        unseen += new_expanding
        
        seen[node] += 1
    
    end = time.time()
    print(end - start)
    return seen
```

# 试试适当加权重，先出来的词权重高，证明效果没什么改善

```python

def get_related_words_add_ratio(initial_words, model):
    
    unseen = [initial_words]
    
    seen = defaultdict(int)
    
    max_size = 400
    
    ratio = 100
    while unseen and len(seen) < max_size:
        if len(seen) % 100 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)    
            
        new_expanding = [w for w, s in model.most_similar(node, topn=20) if is_verb(w)]
        
        unseen += new_expanding
        
        seen[node] += 1 * ratio
        
        if ratio != 1: ratio -= 1
        
    return seen
```

```python
related_words = get_related_words_add_ratio('说', news_word2vec)
```

# 使用PageRank找相近词，证明不行

```python
def get_related_words_by_pagerank(initial_words, model):
    
    unseen = [initial_words]
    
    seen = defaultdict(list)
    
    max_size = 400
    
    while unseen and len(seen) < max_size:
        if len(seen) % 100 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)    
        
        new_expanding = [w for w, s in model.most_similar(node, topn=20) if is_verb(w)]
        
        unseen += new_expanding
        
        for new_node in new_expanding:
#             if new_node not in seen[node]:
            seen[node].append(new_node)
        
    return seen
```

```python
# -*- coding: utf-8 -*-
import os
LTP_DATA_DIR = 'D:\data\ltp_data_v3.4.0'  # ltp模型目录的路径
cws_model_path = os.path.join(LTP_DATA_DIR, 'cws.model')  # 分词模型路径，模型名称为`cws.model`

from pyltp import Segmentor
segmentor = Segmentor()  # 初始化实例
segmentor.load(cws_model_path)  # 加载模型
words = segmentor.segment('在双方首回合的比赛中，苏宁虽然１：２客场告负，但取得了一个宝贵的客场进球，崔龙洙说：“明天的比赛对我个人和球队来讲都是本赛季重要的一场比赛，拥有一个客场进球对我们来说是一个有利的因素，对我们来说还是要从零开始。目前连续的客场作战让队伍比较疲惫，但相信只要我们正常发挥，队员们保持高度专注，打出自己的特点，就一定能获得我们想要的结果，在明天的比赛中最终获胜晋级八强。”\n　　在崔龙洙看来，这场比赛的胜负关键取决于谁能把细节做得更到位，他说：“上港拥有强大的攻击力，而且拥有个人能力突出的球员，对我们来说要做好防守立足不失球，然后在进攻端体现出我们的能力，毕竟我们也拥有进攻方面的实力，明天的比赛全队要保持思想上的高度统一，无论进攻还是防守都要步调一致去执行和落实。”')  # 分词
print('\t'.join(words))
segmentor.release()  # 释放模型
```

```python
pos_model_path = os.path.join(LTP_DATA_DIR, 'pos.model')  # 词性标注模型路径，模型名称为`pos.model`

from pyltp import Postagger
postagger = Postagger() # 初始化实例
postagger.load(pos_model_path)  # 加载模型

postags = postagger.postag(words)  # 词性标注

print('\t'.join(postags))
postagger.release()  # 释放模型
```

```python
ner_model_path = os.path.join(LTP_DATA_DIR, 'ner.model')  # 命名实体识别模型路径，模型名称为`pos.model`

from pyltp import NamedEntityRecognizer
recognizer = NamedEntityRecognizer() # 初始化实例
recognizer.load(ner_model_path)  # 加载模型

netags = recognizer.recognize(words, postags)  # 命名实体识别

print('\t'.join(netags))
recognizer.release()  # 释放模型
```

```python
print(set([word for i, word in enumerate(words) if netags[i] != 'O']))
```

```python
par_model_path = os.path.join(LTP_DATA_DIR, 'parser.model')  # 依存句法分析模型路径，模型名称为`parser.model`

from pyltp import Parser
parser = Parser() # 初始化实例
parser.load(par_model_path)  # 加载模型

arcs = parser.parse(words, postags)  # 句法分析

print("\t".join("%d:%s" % (arc.head, arc.relation) for arc in arcs))
# 利用依存句法提取句子主体
for i, arc in enumerate(arcs):
    if words[arc.head - 1] == '说' and arc.relation == 'SBV':
        print(words[i])
parser.release()  # 释放模型
```

```python
srl_model_path = os.path.join(LTP_DATA_DIR, 'srl')  # 语义角色标注模型目录路径，模型目录为`srl`。注意该模型路径是一个目录，而不是一个文件。

from pyltp import SementicRoleLabeller
labeller = SementicRoleLabeller() # 初始化实例
labeller.load(srl_model_path)  # 加载模型


roles = labeller.label(words, postags, arcs)  # 语义角色标注

# 打印结果
for role in roles:
    print(role.index, "".join(["%s:(%d,%d)" % (arg.name, arg.range.start, arg.range.end) for arg in role.arguments]))
labeller.release()  # 释放模型
```

```python
# 列表中若有元素在object内，返回true，否则返回false
def any_in(list, object):
    for i in list:
        if i in object:
            return True
    return False
```

```python
# 建立词频表
def get_word_frequency(content):
    word_frequency = defaultdict(int)
    for document in content:
        for word in document.split():
            word_frequency[word] += 1
    return word_frequency
```

```python
def idf(word, content, word_frequency):
    """Gets the inversed document frequency"""
    return math.log10(len(content) / word_frequency[word])
```

```python
def tf(word, document):
    """
    Gets the term frequemcy of a @word in a @document.
    """
    words = document.split()
    
    return sum(1 for w in words if w == word)
```

```python
def get_keywords_of_a_document(document, content, word_frequency):
    words = set(document.split())
    
    tfidf = [
        (w, tf(w, document) * idf(w, content, word_frequency)) for w in words
    ]
    
#     tfidf = sorted(tfidf, key=lambda x: x[1], reverse=True)
    
    return tfidf
```

```python
def get_keywords_of_all_document(content, word_frequency):
    tfidf_dict = defaultdict(float)
    for document in content:
        tfidf = get_keywords_of_a_document(document, news_sports_content_cut, word_frequency)
        for word in tfidf:
            tfidf_dict[word[0]] += word[1]
    return tfidf_dict
```

```python
def get_related_words_by_synonyms(initial_word):

    unseen = initial_word.split()
    
    seen = defaultdict(int)
    
    max_size = 400
    
    # 建邻表
    adj = defaultdict(list)
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
        
#         print(node, seen, unseen)
        
        if node in seen:
            new_expanding = adj[node]
        else:            
            new_expanding = [w for w in synonyms.nearby(node)[0]]
        
        adj[node] = new_expanding
        
        unseen += new_expanding
        
        seen[node] += 1

    return seen
```

```python
def get_related_words_by_vec(initial_words, model, topn):

    unseen = initial_words.split()
    
    seen = defaultdict(int)
    
    max_size = 1000
    
    # 建邻表
    adj = defaultdict(list)
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
        
#         print(node, seen)
        
        if node in seen:
            new_expanding = adj[node]
        else:            
            new_expanding = [w for w, s in model.most_similar(node, topn=topn) if is_verb(w)]
        
        adj[node] = new_expanding
        
        unseen += new_expanding
        
        seen[node] += 1
    

    return seen
```
