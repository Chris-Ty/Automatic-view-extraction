```python
from collections import defaultdict
```

```python
# 判断一个词是否是动词，是的话返回True
def is_verb(word):
    for word, flat in pseg.cut(word):
        if flat != 'v': return False        
    return True
```

```python
# 原方法
def get_related_words(initial_words, model, topn):
    start = time.time()
    unseen = initial_words.split()
    
    seen = defaultdict(int)
    
    max_size = 400
    
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
                   
        new_expanding = [w for w, s in model.most_similar(node, topn=topn) if is_verb(w)]
        
        
        unseen += new_expanding
        
        seen[node] += 1
    
    end = time.time()
    print(end - start)
    return seen
```

```python
# 修改后
def get_related_words(initial_words, model, topn):
    start = time.time()
    unseen = initial_words.split()
    
    seen = defaultdict(int)
    
    max_size = 400
    
    # 建邻表
    adj = defaultdict(list)
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
        
#         print(node, seen)
        
        if node in seen:
            new_expanding = adj[node]
        else:            
            new_expanding = [w for w, s in model.most_similar(node, topn=topn) if is_verb(w)]
        
        adj[node] = new_expanding
        
        unseen += new_expanding
        
        seen[node] += 1
    
    end = time.time()
    print(end - start)
    return seen
```

# 试试适当加权重，先出来的词权重高，证明效果没什么改善

```python

def get_related_words_add_ratio(initial_words, model):
    
    unseen = [initial_words]
    
    seen = defaultdict(int)
    
    max_size = 400
    
    ratio = 100
    while unseen and len(seen) < max_size:
        if len(seen) % 100 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)    
            
        new_expanding = [w for w, s in model.most_similar(node, topn=20) if is_verb(w)]
        
        unseen += new_expanding
        
        seen[node] += 1 * ratio
        
        if ratio != 1: ratio -= 1
        
    return seen
```

```python
def get_related_words_by_pagerank(initial_words, model):
    
    unseen = [initial_words]
    
    seen = defaultdict(list)
    
    max_size = 400
    
    while unseen and len(seen) < max_size:
        if len(seen) % 100 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)    
        
        new_expanding = [w for w, s in model.most_similar(node, topn=20) if is_verb(w)]
        
        unseen += new_expanding
        
        for new_node in new_expanding:
#             if new_node not in seen[node]:
            seen[node].append(new_node)
        
    return seen
```

```python
def abstract(view_word, sentence):
    # 分词
    words = segmentor.segment(sentence)
    # 词性标注
    postags = postagger.postag(words)
    # 命名实体识别
    netags = recognizer.recognize(words, postags)
    # 依存句法
    arcs = parser.parse(words, postags)
    
    print(words)
    print(arcs)
    # 提取人物主体，谓语，及观点
    for i, arc in enumerate(arcs):
        # 主语，找出符合条件的主宾关系
        if arc.relation == 'SBV' and words[arc.head - 1] == view_word:
            print('1')
            # 如果主语是代词，需要从前文找出主体
            if postags[i] == 'r':
                sign = 0
                # 向前找主语
                for j in range(i, 0, -1):
                    if netags[j] != 'O':
                        # 找出主体的定语
                        sign = 1
                        if arcs[j].relation == 'ATT':
                            subject = words[j] + words[arcs[j].head - 1]
                        else:
                            subject = words[j]
                        break
                if sign == 0:
                    # 向后找主语
                    for j in range(i, len(words)):
                        if netags[j] != 'O':
                            # 找出主体的定语
                            sign = 1
                            if arcs[j].relation == 'ATT':
                                subject = words[j] + words[arcs[j].head - 1]
                            else:
                                subject = words[j]
                            break
                    # 实在找不到就用代词
                    if sign == 0:
                        subject = words[i]
            else:
                subject = words[i]

            # 谓语
            predicate = view_word

            # 观点
            view = ''

            # 提取观点，先用角色标注，后考虑引号直接拿出，再没有直接将谓语后面的句子输出
            mark = 0

            # 语义角色标注
            roles = labeller.label(words, postags, arcs)

            for role in roles:
                if role.index == arc.head - 1:
                    for arg in role.arguments:
                        if arg.name == 'A1':
                            print('2')
                            # 发现这种方法不准，例如：补充道，会把道作为宾语
                            if arg.range.end - arg.range.start < 2:
                                # 调整谓语
                                view_word += ''.join(words[arg.range.start: arg.range.end + 1])
                                predicate = view_word
                            else:
                                mark = 1
                                view = ''.join(words[arg.range.start: arg.range.end + 1])
                        # 调整谓语，谓语可能有状语,例：我不知道为什么喜欢
                        if arg.name == 'ADV':
                            predicate = ''.join(words[arg.range.start: arg.range.end + 1]) + view_word

            if mark == 0:
                # 引号里面字少的认为不是说的话，可能是特殊词
                pattern = re.compile(r'(“.*?”)')
                if pattern.findall(sentence) != []:
                    print('3')
                    for s in pattern.findall(sentence):
                        if len(s) > 6:
                            view += s
                if view == '':
                    print('4')
                    # 判断是不是标点符号
                    if arc.head < len(words) and postags[arc.head] == 'wp':
                        view = ''.join(words[arc.head + 1:])
                    else:
                        view = ''.join(words[arc.head:])
            return (subject, predicate, view)
    return None
```

```python
# 建立词频表
def get_word_frequency(content):
    word_frequency = defaultdict(int)
    for document in content:
        for word in document.split():
            word_frequency[word] += 1
    return word_frequency
```

```python
def idf(word, content, word_frequency):
    """Gets the inversed document frequency"""
    return math.log10(len(content) / word_frequency[word])
```

```python
def tf(word, document):
    """
    Gets the term frequemcy of a @word in a @document.
    """
    words = document.split()
    
    return sum(1 for w in words if w == word)
```

```python
def get_keywords_of_a_document(document, content, word_frequency):
    words = set(document.split())
    
    tfidf = [
        (w, tf(w, document) * idf(w, content, word_frequency)) for w in words
    ]
    
#     tfidf = sorted(tfidf, key=lambda x: x[1], reverse=True)
    
    return tfidf
```

```python
def get_keywords_of_all_document(content, word_frequency):
    tfidf_dict = defaultdict(float)
    for document in content:
        tfidf = get_keywords_of_a_document(document, news_sports_content_cut, word_frequency)
        for word in tfidf:
            tfidf_dict[word[0]] += word[1]
    return tfidf_dict
```

```python
def get_related_words_by_synonyms(initial_word):

    unseen = initial_word.split()
    
    seen = defaultdict(int)
    
    max_size = 400
    
    # 建邻表
    adj = defaultdict(list)
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0:
            print('seen length:{}'.format(len(seen)))
          
        node = unseen.pop(0)
        
#         print(node, seen, unseen)
        
        if node in seen:
            new_expanding = adj[node]
        else:            
            new_expanding = [w for w in synonyms.nearby(node)[0]]
        
        adj[node] = new_expanding
        
        unseen += new_expanding
        
        seen[node] += 1

    return seen
```
